## Data Visualization

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(),annot =True,cmap = 'pink')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
x=df.groupby(by="Category")[['Profit']].sum()
b=x.sort_values(by=['Profit'],ascending=True)
b.plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Region',y='Profit', hue='Category',data=df)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='Region',y='Sales', hue='Category',data=df)
plt.show()
```

```python
plt.figure(figsize=(8,8))
df.groupby('Region')['Profit'].sum().plot(kind='pie',autopct='%1.1f%%')
plt.show()
```

```python
subcategory_grpby=df.groupby(by="Sub-Category")[['Profit','Sales']].sum()
subprofit=subcategory_grpby.sort_values(by=['Profit'],ascending=True).plot(kind='bar',figsize=(13,8))
```

```python
sns.lineplot(y=df.Profit, x=df.Discount)
plt.show()
```
